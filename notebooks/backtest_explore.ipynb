在项目根目录安装依赖后打开本 Notebook（readme 中有 pip install 说明）。


```python
from trading.baseline_builders import default_baseline_builders_v1
from trading.engine import run_scenarios
from trading.metrics import compare_portfolios
from trading.strategies.dca import DCAParams, fixed_weight_allocator, nasdaq_rule_allocator
from trading.viz import fig_equity_comparison, fig_summary_dashboard

params = DCAParams(
    symbols=("QQQ", "TQQQ"),
    start="2016-01-01",
    end="2026-01-01",
    monthly_budget=5000.0,
    default_weights={"QQQ": 0.7, "TQQQ": 0.3},
    signal_symbol="^IXIC",
    benchmark_symbol="QQQ",
)

results = run_scenarios(
    params,
    allocator=fixed_weight_allocator,
    baseline_builders=default_baseline_builders_v1(params.benchmark_symbol),
)
compare_portfolios(results)
```

```python
fig_summary_dashboard(results).show()
```

```python
fig_equity_comparison(results).show()
```

```python
# 尝试纳指规则分配器
results_rule = run_scenarios(
    params,
    allocator=nasdaq_rule_allocator,
    baseline_builders=default_baseline_builders_v1(params.benchmark_symbol),
)
compare_portfolios(results_rule)
```
